## Filtering reports

```python
from pfd_toolkit import load_reports, LLM, Filter
from dotenv import load_dotenv 
import os

reports = load_reports(start_date="2025-04-01", end_date="2025-05-25")
len(reports)
```

```python
# Load OpenAI API key
load_dotenv("api.env")
openai_api_key = os.getenv("OPENAI_API_KEY")

llm_client = LLM(api_key=openai_api_key, max_workers=30)
```

```python
filter_object = Filter(llm=llm_client,
                        reports = reports,
                        user_query="suicide",
                        filter_df=True)

filtered_reports = filter_object.filter_reports().reset_index(drop=True)
filtered_reports
```
