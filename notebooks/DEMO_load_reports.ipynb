## Loading the PFD reports

The user can load PFD reports without running the scraping module, as the complete dataset is bundled along with the package.

What currently works:

* Importing all PFD reports
* Filtering based on date

What still needs to be developed:

* Filtering based on category
* Currently, the stored reports are HTML+PDF scraped only (not LLM scraped). We will change to only LLM scraped.
* (Optional -- not sure about this yet) A toggle for whether the user wants to import cleaned or uncleaned versions of the report text

#### Get all reports

```python
from pfd_toolkit import load_reports

reports = load_reports()
reports
```

#### See how many reports have been released last year (2024)

```python
reports = load_reports(
    start_date="2025-01-01",
    end_date="2025-05-01")

reports.head()
```

### Read only the 1000 latest reports

```python
reports = load_reports(
    n_reports=1000
)

len(reports)
```
