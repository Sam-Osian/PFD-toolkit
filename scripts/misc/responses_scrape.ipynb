```python
from pfd_toolkit import load_reports, LLM, Scraper
```

```python
reports = load_reports(start_date="2024-11-01", 
                       end_date="2024-11-15",
                       refresh=True)

print(f"{len(reports)} reports loaded.")
```

```python
from dotenv import load_dotenv
import os

# Load OpenAI API key from local environment
load_dotenv("api.env")
openai_api_key = os.getenv("OPENAI_API_KEY")

# Initialise LLM client
llm_client = LLM(api_key=openai_api_key, 
                 max_workers=30, 
                 model="gpt-4.1-mini",
                 seed=123, 
                 temperature=0, 
                 timeout=60)
```

```python
scraper = Scraper(llm=llm_client)

responses = scraper.scrape_responses(reports=reports)

responses
```
